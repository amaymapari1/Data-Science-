```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset into a pandas DataFrame
df = pd.read_csv('Iris.csv')

# 2. List down the features and their types
feature_types = df.dtypes
print(feature_types)

# 3. Create a histogram for each feature
df.hist(figsize=(10, 10))
plt.show()

# 4. Create a box plot for each feature
df.boxplot(figsize=(10, 6))
plt.show()

# Identify outliers using numerical columns
numerical_columns = df.select_dtypes(include=['int', 'float']).columns
for column in numerical_columns:
    Q1=np.percentile(df[column],25,interpolation='midpoint')
    Q3=np.percentile(df[column],75,interpolation='midpoint')
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers in {column}:")
    print(outliers)

# You can also compare feature distributions visually using other plots like KDE plots or violin plots
```

```python
df1 = sns.load_dataset('titanic')

# 1. Plot a box plot for distribution of age with respect to each gender and survival
sns.boxplot(x='sex', y='age', hue='survived', data=df1)
plt.title('Distribution of Age with Respect to Gender and Survival')
# plt.show
```

```python
sns.catplot(x='sex',y='age', hue='survived', data=df1)
plt.title('Passenger Count by Gender and Survival')
plt.show()
```

```python
sns.violinplot(x='sex',y='age', hue='survived', data=df1)
plt.title('Passenger Count by Gender and Survival')
plt.show()
```

```python
sns.scatterplot(x='sex',y='age', hue='survived', data=df1)
plt.title('Passenger Count by Gender and Survival')
plt.show()
```

```python
sns.swarmplot(x='sex',y='age', hue='survived', data=df1)
plt.title('Passenger Count by Gender and Survival')
plt.show()
```

```python
sns.stripplot(x='sex',y='age', hue='survived', data=df1)
plt.title('Passenger Count by Gender and Survival')
plt.show()
```
