```python
# #zscore method to remove outliers
outliers=[]
def detect_outliers(x):
    threshold=1
    m=x.mean()
    s=x.std()
    
    for i in x:
        z_score=(i-m)/s
#         print(z_score)
        if np.abs(z_score)>threshold:
            outliers.append(i)
#             x(i).drop()
            
    return outliers        
threshold=1
m=df['math score'].mean()
s=df['math score'].std()
zscores=(df['math score']-m)/s
count=0
for i in zscores:
    if i>threshold:
        count+=1
print(count)
print(df.shape)
df2=df[zscores <threshold]
print(df2.shape)
df2
```

```python
outliers=detect_outliers(df['math score'])
len(outliers)
```
